```python
difficulty_Cols = [
    "Read_diff", 
    "Write_diff",
    "Conv_diff",
    "Lect_diff",
    "Speak_diff"]

df["Learning_difficulty_index"]= df[difficulty_Cols].mean(axis=1)
```

SKill Gap analysis 

```python
df["Read_skill"]  = 5 - df["Read_diff"]
df["Write_skill"] = 5 - df["Write_diff"]
df["Speak_skill"] = 5 - df["Speak_diff"]
df["Lect_skill"]  = 5- df["Lect_diff"]
df["conv_skill"]  = 5 - df["Conv_diff"]
```

```python
skill_cols = ["Read_skill", "Write_skill", "Speak_skill", "Lect_skill","conv_skill"]

expected = df[skill_cols].quantile(0.75).mean()

for col in skill_cols:
    df[col + "_gap"] = expected - df[col]
```

```python
expected
```

```python
gap_cols = [col + "_gap" for col in skill_cols]

df["core_Skill_Gap"] = df[gap_cols].mean(axis=1)
```

```python
def categorize_gap(x):
    if x > 2:
        return "High Gap"
    elif x > 1:
        return "Medium Gap"
    else:
        return "Low Gap"

df["Gap_Level"] = df["core_Skill_Gap"].apply(categorize_gap)
```

Clustering of skill gap

```python
features = [
    "Read_skill_gap",
    "Write_skill_gap",
    "Speak_skill_gap",
    "Lect_skill_gap",
    "conv_skill_gap"
]
```

```python
X = df[features].dropna()
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans


inertia =[]

K_range= range(1,10)

for K in K_range:
    kMeans = KMeans(n_clusters= K, random_state = 42, n_init= 10)
    kMeans.fit(X_scaled)
    inertia.append(kMeans.inertia_)
```

```python
import matplotlib.pyplot as plt

plt.plot(K_range, inertia, marker='o')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("inertia")
plt.title("Elbow method")
plt.show()
```

```python
kmeans = KMeans(n_clusters= 3, random_state=42)
df["cluster"]= kmeans.fit_predict(X_scaled)
```

```python
df["cluster"].value_counts()
```

```python
df.groupby("cluster")[features].mean()
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components =2, random_state = 42)
X_tsne = tsne.fit_transform(X_scaled)

plt.figure(figsize=(8,8))

plt.scatter(
    X_tsne[:,0],
    X_tsne[:,1],
    c= df["cluster"],
    cmap= "Set1",
    s=50
)

plt.title("cluster")
plt.colorbar(label= "cluster")
plt.grid(True)
plt.show
```
