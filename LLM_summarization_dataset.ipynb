```python
core_skills = [
    "Conv_diff",   
    "Speak_diff",
    "Write_diff",
    "Read_diff",
    "Lect_diff"
]
```

```python
summary = df[core_skills].mean().sort_values(ascending=False)
print(summary)
```

```python
text_data = "The analysis of student difficulties in English language skills shows that "

for i, (col, val) in enumerate(summary.items()):
    skill = col.replace("_diff", "").capitalize()
    
    if i == 0:
        text_data += f"{skill} is the most difficult skill with an average score of {val:.2f}. "
    else:
        text_data += f"{skill} has an average difficulty score of {val:.2f}. "

print(text_data)
```

```python
prompt = f"""
Analyze the following data:

{text_data}


1. Rank the skills from most to least difficult.
2. What overall learning pattern do you observe?
3. Do students struggle more with communication or comprehension skills?
4. What major challenges do students face?
5. What factors contribute to these difficulties?
6. Provide 2–3 key insights.
7. Identify any surprising trends.
8. What improvements would you suggest for students?
"""
```
