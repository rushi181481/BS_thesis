Age group vs Skill difficulty


```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f_oneway

cols = [
    'Conv_diff', 'Speak_diff', 'Write_diff', 'Read_diff', 'MT_influence',
    'Lect_diff', 'Movie_Nosub', 'Movie_sub', 'Learning_feel', 'Switch_diff',
    'work_india', 'Work_abroad', 'Study_other', 'Friend_diff'
]

# Mean table by age group
avg_diff_age = df.groupby('Age group')[cols].mean()
print("\nAverage Difficulty by Age Group:\n")
print(avg_diff_age)
```

```python
for col in cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x='Age group', y=col, data=df)
    plt.title(f"{col} by Age Group")
    plt.ylabel("Difficulty Score")
    plt.xlabel("Age Group")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()


age_groups = df['Age group'].dropna().unique()

for col in cols:
    group_vals = [df[df['Age group'] == grp][col].dropna() for grp in age_groups]
    
    
    group_vals = [g for g in group_vals if len(g) > 0]
    
    if len(group_vals) >= 2:
        f_stat, p_val = f_oneway(*group_vals)
        print(f"{col}: F-statistic = {f_stat:.3f}, p-value = {p_val:.3f}")
        
        if p_val < 0.05:
            print(f"→ Significant difference exists among the age groups for {col}.\n")
        else:
            print(f"→ No significant difference among the age groups for {col}.\n")
    else:
        print(f"{col}: Not enough groups available for ANOVA.\n")
```

Gender vs skill difficulties

```python
import pandas as pd
from scipy.stats import ttest_ind


difficulty_cols = [
    'Conv_diff', 'Speak_diff', 'Write_diff', 'Read_diff',
    'Lect_diff', 'Movie_Nosub', 'Movie_sub',
    'Learning_feel', 'Switch_diff',
    'work_india', 'Work_abroad', 'Study_other',
    'Friend_diff', 'MT_influence'
]

results = []

for col in difficulty_cols:
    male = df[df['Gender'] == 'Male'][col].dropna()
    female = df[df['Gender'] == 'Female'][col].dropna()
    
    t_stat, p_val = ttest_ind(male, female, equal_var=False)  
    
    results.append({
        'Variable': col,
        'Male_Mean': male.mean(),
        'Female_Mean': female.mean(),
        't_statistic': round(t_stat, 3),
        'p_value': round(p_val, 4),
        'Significant (p<0.05)': 'Yes' if p_val < 0.05 else 'No'
    })


results_df = pd.DataFrame(results)

print(results_df)

import seaborn as sns
import matplotlib.pyplot as plt

for col in difficulty_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x='Gender', y=col, data=df)
    plt.title(f"{col} by Gender")
    plt.show()

```

Medium vs Skill difficulties

```python
all_cols = ['Conv_diff', 'Speak_diff', 'Write_diff', 'Read_diff', 
            'Lect_diff', 'Movie_Nosub', 'Movie_sub', 'Learning_feel',
            'Switch_diff', 'work_india', 'Work_abroad', 'Study_other',
            'Friend_diff', 'MT_influence']

avg_diff_all = df.groupby('Medium of School ')[all_cols].mean()
print("\nAverage Difficulty by Medium of School:\n", avg_diff_all)


for col in all_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x='Medium of School ', y=col, data=df)
    plt.title(f"{col} by Medium of School")
    plt.ylabel("Difficulty Score")
    plt.xlabel("Medium of School")
    plt.tight_layout()
    plt.show()

from scipy.stats import f_oneway

for col in all_cols:
    telugu_vals = df[df['Medium of School ']=='Telugu medium'][col].dropna()
    english_vals = df[df['Medium of School ']=='English medium'][col].dropna()
    both_vals = df[df['Medium of School ']=='Both'][col].dropna()
    
    f_stat, p_val = f_oneway(telugu_vals, english_vals, both_vals)
    print(f"{col}: F-statistic = {f_stat:.3f}, p-value = {p_val:.3f}")
    
    if p_val < 0.05:
        print(f"→ Significant difference exists among the three groups for {col}.\n")
    else:
        print(f"→ No significant difference among the three groups for {col}.\n")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_rel

print("Average Difficulty Scores")

print("Movies without subtitles:", df['Movie_Nosub'].mean())
print("Movies with subtitles:", df['Movie_sub'].mean())


t_stat, p_value = ttest_rel(df['Movie_Nosub'], df['Movie_sub'])

print("\nPaired T-Test Results")
print("T-statistic:", t_stat)
print("P-value:", p_value)


data = pd.DataFrame({
    'Without Subtitles': df['Movie_Nosub'],
    'With Subtitles': df['Movie_sub']
})

plt.figure(figsize=(6,5))
sns.boxplot(data=data)

plt.title("Difficulty Comparison: Movies With vs Without Subtitles")
plt.ylabel("Difficulty Score")
plt.show()
```
