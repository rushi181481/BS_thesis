Gender vs skill difficulties

```python
import pandas as pd
from scipy.stats import ttest_ind

# List all difficulty-related columns
difficulty_cols = [
    'Conv_diff', 'Speak_diff', 'Write_diff', 'Read_diff',
    'Lect_diff', 'Movie_Nosub', 'Movie_sub',
    'Learning_feel', 'Switch_diff',
    'work_india', 'Work_abroad', 'Study_other',
    'Friend_diff', 'MT_influence'
]

results = []

for col in difficulty_cols:
    male = df[df['Gender'] == 'Male'][col].dropna()
    female = df[df['Gender'] == 'Female'][col].dropna()
    
    t_stat, p_val = ttest_ind(male, female, equal_var=False)  # Welch's test
    
    results.append({
        'Variable': col,
        'Male_Mean': male.mean(),
        'Female_Mean': female.mean(),
        't_statistic': round(t_stat, 3),
        'p_value': round(p_val, 4),
        'Significant (p<0.05)': 'Yes' if p_val < 0.05 else 'No'
    })

# Convert to DataFrame
results_df = pd.DataFrame(results)

print(results_df)

import seaborn as sns
import matplotlib.pyplot as plt

for col in difficulty_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x='Gender', y=col, data=df)
    plt.title(f"{col} by Gender")
    plt.show()

```

Medium vs Skill difficulties

```python
# Filter only English medium and Telugu medium students
df_filtered = df[df['Medium of School '].isin(['English medium', 'Telugu medium'])]

# Columns to analyze
cols = ['Speak_diff', 'Lect_diff', 'Switch_diff', 'MT_influence']

#  Average difficulty by Medium of School
avg_diff_filtered = df_filtered.groupby('Medium of School ')[cols].mean()
print("\nAverage Difficulty (English vs Telugu medium):\n", avg_diff_filtered)

#  Boxplots
plt.figure(figsize=(12,8))
for i, col in enumerate(cols):
    plt.subplot(2,2,i+1)
    sns.boxplot(x='Medium of School ', y=col, data=df_filtered)
    plt.title(f"{col} by Medium of School")
    plt.ylabel("Difficulty Score")
plt.tight_layout()
plt.show()

#  T-tests Welch's
for col in cols:
    telugu_vals = df_filtered[df_filtered['Medium of School ']=='Telugu medium'][col].dropna()
    english_vals = df_filtered[df_filtered['Medium of School ']=='English medium'][col].dropna()
    
    t_stat, p_val = ttest_ind(telugu_vals, english_vals, equal_var=False)
    print(f"{col}: t-statistic = {t_stat:.3f}, p-value = {p_val:.3f}")
    
    if p_val < 0.05:
        print(f"→ Significant difference: Telugu medium students have higher difficulty in {col}.\n")
    else:
        print(f"→ No significant difference in {col}.\n")
```

```python
all_cols = ['Conv_diff', 'Speak_diff', 'Write_diff', 'Read_diff', 
            'Lect_diff', 'Movie_Nosub', 'Movie_sub', 'Learning_feel',
            'Switch_diff', 'work_india', 'Work_abroad', 'Study_other',
            'Friend_diff', 'MT_influence']

# Average difficulty by Medium of School including all students (both)
avg_diff_all = df.groupby('Medium of School ')[all_cols].mean()
print("\nAverage Difficulty by Medium of School:\n", avg_diff_all)


for col in all_cols:
    plt.figure(figsize=(6,4))   # smaller size
    sns.boxplot(x='Medium of School ', y=col, data=df)
    plt.title(f"{col} by Medium of School")
    plt.ylabel("Difficulty Score")
    plt.xlabel("Medium of School")
    plt.tight_layout()
    plt.show()

from scipy.stats import f_oneway

for col in all_cols:
    telugu_vals = df[df['Medium of School ']=='Telugu medium'][col].dropna()
    english_vals = df[df['Medium of School ']=='English medium'][col].dropna()
    both_vals = df[df['Medium of School ']=='Both'][col].dropna()
    
    f_stat, p_val = f_oneway(telugu_vals, english_vals, both_vals)
    print(f"{col}: F-statistic = {f_stat:.3f}, p-value = {p_val:.3f}")
    
    if p_val < 0.05:
        print(f"→ Significant difference exists among the three groups for {col}.\n")
    else:
        print(f"→ No significant difference among the three groups for {col}.\n")
```
